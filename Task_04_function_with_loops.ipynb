```python
# =====================================================
# Problem 1: Employee Performance Bonus Eligibility
# =====================================================

employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Find highest score
max_score = max(employees.values())

# Find all employees with highest score (handle ties)
top_performers = [name for name, score in employees.items() if score == max_score]

print("Top Performers Eligible for Bonus:", ", ".join(top_performers), f"(Score: {max_score})")
```

```python
# =====================================================
# Problem 2: Search Query Keyword Analysis
# =====================================================

query = "Buy mobile phone buy phone online"

# Normalize text
query = query.lower()

# Remove punctuation (basic)
import string
query = query.translate(str.maketrans("", "", string.punctuation))

words = query.split()

# Count frequency
freq = {}
for word in words:
    freq[word] = freq.get(word, 0) + 1

# Keep only repeated keywords
repeated_keywords = {k: v for k, v in freq.items() if v > 1}

print(repeated_keywords)
```

```python
# =====================================================
# Problem 3: Sensor Data Validation
# =====================================================

sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

for hour, value in enumerate(sensor_readings):
    if value % 2 == 0:  # even readings are valid
        valid_readings.append((hour, value))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
# =====================================================
# Problem 4: Email Domain Usage Analysis
# =====================================================

emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}

for email in emails:
    domain = email.split("@")[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1

total = len(emails)

for domain, count in domain_count.items():
    percentage = (count / total) * 100
    print(f"{domain}: {percentage:.0f}%")
```

```python
# =====================================================
# Problem 5: Sales Spike Detection
# =====================================================

sales = [1200, 1500, 900, 2200, 1400, 3000]

avg_sales = sum(sales) / len(sales)
threshold = avg_sales * 1.3  # 30% above average

for day, value in enumerate(sales, start=1):
    if value > threshold:
        print(f"Day {day}: {value}")
```

```python
# =====================================================
# Problem 6: Duplicate User ID Detection
# =====================================================

user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

count = {}

for uid in user_ids:
    count[uid] = count.get(uid, 0) + 1

# Show only duplicates
for uid, c in count.items():
    if c > 1:
        print(f"{uid} → {c} times")
```
