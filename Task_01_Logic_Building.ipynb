# Task 1: User Login Check

Description: Validate username and password

Real-World Use: Authentication Systems

```python
# Given credentials
username = "admin"
password = "1234"

# Login validation logic
if username == "admin" and password == "1234":
    print("Login Successful")
else:
    print("Invalid Credentials")
```

# Task 2: Pass / Fail Analyzer

Description: Analyze student performance

Real-World Use: Academic Result Systems

```python
marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

# Iterate through marks
for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

# Display results
print("Total Students :", len(marks))
print("Passed Students :", pass_count)
print("Failed Students :", fail_count)
```

# Task 3: Simple Data Cleaner

Description: Clean and standardize user names

Real-World Use: Data Preprocessing

```python
names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []

# Data cleaning process
for name in names:
    cleaned_name = name.strip().lower()
    cleaned_names.append(cleaned_name)

print("Cleaned Names :", cleaned_names)
```

# Task 4: Message Length Analyzer

Description: Analyze message size

Real-World Use: Text Validation Systems

```python
messages = ["Hi", "Welcome to the platform", "OK"]

for message in messages:
    length = len(message)
    print(f"Message: '{message}' | Length: {length}")

    if length > 10:
        print("Status: Long Message")
```

# Task 5: Error Message Detector

Description: Detect and count system errors

Real-World Use: Log Monitoring Systems

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = 0

# Count ERROR entries
for log in logs:
    if log == "ERROR":
        error_count += 1

print("Total ERROR Messages :", error_count)
```
